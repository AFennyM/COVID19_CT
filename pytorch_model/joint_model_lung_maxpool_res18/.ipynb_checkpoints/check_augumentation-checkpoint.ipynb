```python
df = pd.read_csv('/data/COVID19_clinical_data/clinical_info.csv')
feature_list = ['Sex', 'Age', 'Exposure', 'ClinicalSuspicious',
       'ImagingSuspicious', 'Fever', 'Sputum', 'Cough', 'Temperature',
       'WhiteBloodCell', 'Neutrophil', 'NeutrophilPercent', 'Lymphocyte',
       'LymphocytePercent']

pid2features = {}
for p, v in zip(df.pid, df[feature_list].values):
    pid2features[p] = v
    
train_info, val_info = get_train_split(0)
train_info = [i for i in train_info if i in pid2features]
val_info = [i for i in val_info if i in pid2features]
```
